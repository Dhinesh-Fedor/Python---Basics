```python
#Pandas Panel  is a 3D type of the container
import pandas as pd
df=pd.DataFrame({'Animal':['Falcon','Falcon','Parrot','Parrot'],'Max Speed': [380.,370.,24.,26.]})
df.groupby("Animal",group_keys=True)[['Max Speed']].apply(lambda x: x)
```

```python
import pandas as pd
import numpy as np
data=np.random.rand(1,3,2)
print("data:\n",data)
print()
df=pd.DataFrame(data.reshape(3,2),index=pd.MultiIndex.from_tuples([(0,i) for i in range(3)],name=['level_0','level_1']),columns=['A','B'])
print('DataFrame:\n',df)
```

```python
rng=np.random.RandomState(0)
ds = pd.DataFrame({'key':['A','B','C','A','B','C'],'data1': range(6),'data2':rng.randint(0,10,6)},columns = ['key','data1','data2'])
ds
```

```python
ds.groupby('key').aggregate(['min',np.median,max])
```

```python
def filter_func(x):
   return x['data2'].std() > 4

display('ds',ds.groupby('key').std(),ds.groupby('key').filter(filter_func))
```

```python
rng=np.random.RandomState(0)
da = pd.DataFrame({'key':['A','B','C','A','B','C'],'data1': range(6),'data2':rng.randint(0,10,6)},columns = ['key','data1','data2'])
da
```

```python
def center(x):
  return x-x.mean()

da.groupby('key').transform(center)
```

```python
da['size'] = da.groupby('key')['data1'].transform(len)
da
```

```python
def norm_by_data2(x):
  x['data1']/=x['data2'].sum()
  return x

da.groupby('key').apply(norm_by_data2)
```

```python
L=[0,1,0,1,2,0]
da.groupby(L).sum()
```
