```python
#Pandas Objects: Concatenation , Merging and Joining  DataFrames , Index Objects

import pandas as pd
import numpy as np


#Function to create DataFrames
def make_df(cols,index):
  data = { c: [str(c) + str(i) for i in index] for c in cols}
  return pd.DataFrame(data,index)

df = make_df('ABC',range(3))
df

```

```python
#Concatenation of NumPy Arrays

x = [1,2,3]
y = [4,5,6]
z = [7,8,9]

np.concatenate([x,y,z])
```

```python
x = [
    [1,2],
    [3,4],
]

np.concatenate([x,x],axis=1)

#Note:- by default axis is  0 (ie) row
```

```python
#Concatenation of Panda

#Syntax => pd.concat()

a = pd.Series(['A','B','C'])
b = pd.Series(['D','E','F'])

pd.concat([a,b],axis = 0)

#Note:- by default axis is 0 (ie) rows
```

```python
df1 = make_df('AB',[1,2])
df2 = make_df('AB',[3,4])

disp = df1,df2,pd.concat([df1,df2])

#Using display method from IPython.display to display the objects  (Another way to print)

#Importing display
from IPython.display import display
display(*disp)
```

```python
#Concatenating a dataframe with series

#dataframe
data = {
    'Name' : ['A','B','C'],
    'Age' : [1,2,3],
    'job' : ['no job','job1','job2']
}

df = pd.DataFrame(data)
df
```

```python
#Series
s = pd.Series(['no salary',2,3], name= 'Salary')
s
```

```python
#Concatenation
pd.concat([df,s],axis=1)
```

```python
#Concatenation with joins

#Three types of joins => Outer, Inner, Axes
#Syntax => join = 'outer', join = 'inner',axis = 0 or 1 (row or column)

df5 = make_df('ABC',[1,2])
df6 = make_df('BCD',[3,4])

#Outer join
display(pd.concat([df5,df6],join = 'outer'))
```

```python
#Inner join
display(pd.concat([df5,df6],join = 'inner'))
```

```python
#Axes
display(pd.concat([df5,df6],axis = 1))
```

```python
#Reindexing
display(pd.concat([df5,df6.reindex(df5.columns,axis = 1)]))
```

```python
#Joining DataFrames

data = {
    'Name' : ['A','B','C'],
    'Age' : [1,2,3],
    'job' : ['no job','job1','job2']
}

df7 = pd.DataFrame(data)
df7

df8 = pd.DataFrame({
    'Address': ['AA','BB','CC']
})

res = df7.join(df8)
display(res)
```

```python
df7.join(df8, on = 'Age')
```

```python
df7.join(df8, on = 'Age',how = 'inner')
```

```python
#Merge Operation on DataFrame

data = {
    'Name' : ['AA','AB','AC'],
    'Age' : [1,2,3],
    'job' : ['no job','job1','job2']
}

df7 = pd.DataFrame(data)
display(df7)

data = {
    'Name' : ['A','B','C'],
    'Age' : [1,2,3],
    'job' : ['no job','job1','job2']
}

df9 = pd.DataFrame(data)
display(df9)

```

```python
#Merge
res = pd.merge(df7,df9,on = 'Age')
res
```

```python
#Merging using keys
res1 = pd.merge(df7,df9,how = 'right',on = ['Name','Age'])
res1
```

```python
#Another way of Merge

dff = make_df('ABC',range(3))
dff1 = make_df('CDE',range(3))

merge = {'data1' : dff, 'data2': dff1}
pd.concat(merge)
```
