```python
#Operating on Null values

#isnull => returns boolean value on whether null is present or not
#notnull => opposite of isnull
#dropna => filters the data by removing null
#fillna => fills na with some value
```

```python
h =  pd.Series([1,np.nan,None,pd.NA,3,4,'bob'])
h
```

```python
h.isnull()
```

```python
h.notnull()
```

```python
h.dropna()
```

```python
df = pd.DataFrame([
    [1,np.nan,3],
    [4,5,6],
    [np.nan,8,9],
])

df
```

```python
#column wise dropping
df.dropna(axis=1)
```

```python
#row wise (default)
df.dropna()
```

```python
df.isnull()
```

```python
df[3] = np.nan
df
```

```python
df.dropna(axis=1,how = 'all')
```

```python
df.dropna(thresh=3) #prints row 2 (has 3 not null values)

#Note:- checks for number of not null values
```

```python
df.dropna(axis=1,thresh=3)
```

```python
#Filling NUll Values

i =  pd.Series([1,np.nan,None,3,4],index = list('abcde'))
i
```

```python
#Forward fill

i.fillna(method='ffill')
```

```python
#Back fill

i.fillna(method='bfill')
```

```python
#column wise forward fill
df.fillna(axis=1,method='ffill')
```

```python
#column wise backward fill
df.fillna(axis=1,method='bfill')
```

```python
#rowwise forward fill
df.fillna(method='ffill')
```

```python
#rowwise backward fill
df.fillna(method='bfill')
```
