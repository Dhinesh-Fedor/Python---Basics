```python
#Modifying a dataframe

data = [
    { 'name' : 'Asabeneh', 'country' : 'Finland', 'city' : 'Helsink'},
    { 'name' : 'David', 'country' : 'UK', 'city' : 'London'},
    { 'name' : 'John', 'country' : 'Sweden', 'city' : 'Stockholm'},

]

data_frame = pd.DataFrame(data)
print(data_frame)


data_frame.columns
```

```python
#Adding a new column


weight = [50,60,70]
data_frame['weight'] = weight

height = [150,160,170]
data_frame['height'] = height



data_frame
```

```python
#Deleting a label from the dataframe

delete = [1,2,3]
data_frame['delete'] = delete
print(data_frame)

data_frame.drop(labels = 'delete',axis = 1)
```

```python
#Modifying column values


#Calculating bmi
def BMI():
  weight = data_frame['weight']
  height = data_frame['height']
  return weight/(height**2)

res = BMI()


#adding it to the dataframe

data_frame['bmi'] = res
data_frame



```

```python
#Formatting dataframe columns

#round() method is used

data_frame['bmi'] = round(data_frame['bmi'],1)
data_frame
```

```python
#Adding 2 more columns

birth_year = [1990,1999,2000]
data_frame['birth_year'] = birth_year

current_year = pd.Series(2024,index = [0,1,2])
data_frame['current_year'] = current_year
data_frame

age = data_frame['current_year'] - data_frame['birth_year']
data_frame['age'] = age
data_frame
```

```python
#checking datatype of a label
print(data_frame.age.dtype)


#Convertig types
data_frame.age.astype(float)
```

```python
#Boolean indexing
#Syntax => dataframe[condition]

data_frame[data_frame['age'] > 25]
```
