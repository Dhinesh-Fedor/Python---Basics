```python
#Data Wrangling With Pandas

#Data Exploration

import pandas as pd

data = {
    'Name': ['A','B','C','D'],
    'Gender' : ['M','F','F','M'],
    'Age' : [1,2,3,4],
    'Marks': [20,'NaN',40,'NaN']
}

df = pd.DataFrame(data)
df
```

```python
#Dealing with missing values in python

sum = 0
avg = 0
NaN =  0

for i in df['Marks']:
  if str(i).isnumeric():
    sum += i
  else:
     NaN +=1
avg = sum/NaN

df = df.replace(to_replace = 'NaN',value = avg)
df
```

```python
df['Gender'] = df['Gender'].map({
      'M' : 0,
      'F' : 1,
}).astype(float)

df
```

```python
#Pandas Reshape

#Reshape using pivot() => reshapes based on column values


data = {
    'Date' : ['2023-01-01','2023-09-08','2023-05-07','2023-02-05'],
    'Category' : ['A','B','C','D'],
    'Value' : [10,20,30,40]
}

df1 = pd.DataFrame(data)
df1
```

```python
pivot_df1 = df1.pivot(index = 'Date', columns = 'Category', values = 'Value',)
print(pivot_df1)
```

```python
pivot_df = df.pivot(index = 'Name', columns = 'Gender', values = 'Marks',)
print(pivot_df)
```

```python
#Reshaping with pivot_table using aggregate function (aggfunc)

pivot_df1 = df1.pivot_table(index = 'Category',  values = 'Value', aggfunc = 'mean')
print(pivot_df1)
```

```python
#Filtering Data in Data Wrangling

#filtering data based on marks
df4 = df[df['Marks'] > 20].copy()

#Deleting the age column
df4.drop('Age',axis = 1, inplace = True)

df4
```

```python
#Merging Dataframe

value = [10,20,30,40]
df['Value'] = value
df

merge_df = pd.merge(df,df1, on= 'Value')
merge_df
```

```python
#Grouping dataframe

group_df = df.groupby('Marks')
print(group_df)

for i in group_df:
  print(i)
```

```python
#Data Wrangling by Renoving Duplication

#Syntax => DataFrame.duplicated(subset = ___ , keep '__')
#returns bool value

#checking duplilcate
df.duplicated(subset = 'Gender', keep = 'first')
```

```python
#Non - duplicates
non_dup = df[~df.duplicated('Gender')]
non_dup
```
