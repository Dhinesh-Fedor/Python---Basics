```python
#Common Plots for statistical analysis

import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as pex


#Bar Graph

x = [10,20,30,40,50,60]
y = [13,45,23,34,96,76]

plt.title('BAR GRAPH')
plt.bar(x,y,color= 'blue', width =4)
plt.show()
```

```python
#Pie Chart

x = [10,15,30,40,5,45]
y = ['a','b','c','d','e','f']

plt.title('PIE GRAPH')
plt.pie(x,labels = y)
plt.show()
```

```python
#Line Plot
data = sb.load_dataset('iris')
sb.lineplot(x="sepal_length",y="sepal_width",data=data)
```

```python
#Scatter Plot

sb.scatterplot(data = data)
```

```python
#Box Plot

df = pex.data.iris()
plot = pex.box(df,x = "sepal_width",y = "sepal_length")
plot.show()
```

```python
#Histogram

df = pex.data.tips()
hist = pex.histogram(df.total_bill)
hist.show()
```

```python
#Plotnine (ggplot in Python)

from plotnine import ggplot, aes, geom_point
import pandas as pd

df = pd.DataFrame({
    'x' : range(10),
    'y' : range(10),
})


plot = ggplot(df,aes('x','y')) + geom_point()
plot
```

```python
#Scatter Plot using Matplotlib

x = range(10)
y = range(10)

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Scatter Plot using Bokeh

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

df = pd.DataFrame({
    'x' : range(10),
    'y' : range(10),
})


plot = figure(title = "Scatter Plot",x_axis_label = 'x',y_axis_label = 'y')
plot.scatter(df['x'],df['y'], size = 10, color = 'black', alpha = 0.5)

show(plot)
```

```python
#Line Plot using plotnine

from plotnine import geom_line

plot = ggplot(df, aes('x','y'))+geom_line()
plot
```

```python
#Line Plot using  Matplotlib

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Lineplot using Bokeh

import numpy as np

output_notebook()


df = pd.DataFrame({
    'x': range(10),
    'y' : np.random.randn(10)
})


plot = figure(title = "Line Plot",x_axis_label = 'x',y_axis_label = 'y')
plot.line(df['x'],df['y'],color = 'black', line_width = 2)

show(plot)
```

```python
#Histogram using plotnine

from plotnine import geom_histogram

df = pd.DataFrame({
    'x' : [1,1,1,2,3,4,10,22,34]
})


plot = ggplot(df,aes('x')) + geom_histogram(binwidth = 2)
plot


```

```python
#Histogram using Bokeh

output_notebook()


data = np.random.randn(1000)

hist, edges = np.histogram(data,bins = 10)

plot = figure(title = "Line Plot",x_axis_label = 'x',y_axis_label = 'y')
plot.quad(top = hist, bottom = 0, left = edges[:-1],right = edges[1:], fill_color="blue",line_color = "red",alpha = 0.5)

show(plot)

```

```python
#Histogram using Matplotlib

plt.hist(data,bins = 10)
plt.title("Histogram")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

```
