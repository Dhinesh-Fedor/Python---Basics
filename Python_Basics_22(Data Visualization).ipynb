```python
#Common Plots for statistical analysis

import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as pex


#Bar Graph

x = [10,20,30,40,50,60]
y = [13,45,23,34,96,76]

plt.title('BAR GRAPH')
plt.bar(x,y,color= 'blue', width =4)
plt.show()
```

```python
#Pie Chart

x = [10,15,30,40,5,45]
y = ['a','b','c','d','e','f']

plt.title('PIE GRAPH')
plt.pie(x,labels = y)
plt.show()
```

```python
#Plotnine (ggplot in Python)

from plotnine import ggplot, aes, geom_point
import pandas as pd

df = pd.DataFrame({
    'x' : range(10),
    'y' : range(10),
})


plot = ggplot(df,aes('x','y')) + geom_point()
plot
```

```python
#Scatter Plot using Matplotlib

x = range(10)
y = range(10)

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Scatter Plot using Bokeh

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

df = pd.DataFrame({
    'x' : range(10),
    'y' : range(10),
})


plot = figure(title = "Scatter Plot",x_axis_label = 'x',y_axis_label = 'y')
plot.scatter(df['x'],df['y'], size = 10, color = 'black', alpha = 0.5)

show(plot)
```

```python
#Line Plot using plotnine

from plotnine import geom_line

plot = ggplot(df, aes('x','y'))+geom_line()
plot
```

```python
#Line Plot using  Matplotlib

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Lineplot using Bokeh

import numpy as np

output_notebook()


df = pd.DataFrame({
    'x': range(10),
    'y' : np.random.randn(10)
})


plot = figure(title = "Line Plot",x_axis_label = 'x',y_axis_label = 'y')
plot.line(df['x'],df['y'],color = 'black', line_width = 2)

show(plot)
```

```python
#Histogram using plotnine

from plotnine import geom_histogram

df = pd.DataFrame({
    'x' : [1,2,2,3,3,3,4,4,4,4]
})


plot = ggplot(df,aes('x')) + geom_histogram(binwidth = 1)

plot


```

```python
#Histogram using Bokeh

output_notebook()


data = np.random.randn(1000)

hist, edges = np.histogram(data,bins = 30)

plot = figure(title = "Histogram",x_axis_label = 'x',y_axis_label = 'y')
plot.quad(top = hist, bottom = 0, left = edges[:-1],right = edges[1:], fill_color="blue",line_color = "red",alpha = 0.5)

show(plot)

```

```python
#Histogram using Matplotlib

plt.hist(data,bins = 30)
plt.title("Histogram")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

```

```python
#Box Plot using Plotnine


from plotnine import geom_boxplot

df = pd.DataFrame({
    'x' : ['A','B','C','D'],
    'y' : [1,2,3,4],

})


plot = ggplot(df,aes('x','y')) + geom_boxplot()

plot
```

```python
#Boxplot using Bokeh

output_notebook()


x  = ['A','B','C','D']
y = np.random.randn(10,4)



plot = figure(title = "Box Plot",x_range = x)

for i, x in enumerate(x):
  plot.vbar(x = [x]*len(y),top = y[:,1], width = 0.3)

show(plot)
```

```python
#BoxPlot using Matplotlib

data = [np.random.randn(50),np.random.randn(50),np.random.randn(50),]

plt.boxplot(data,labels = ['A','B','C'])
plt.title("Box Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Bargragh using Plotnine


from plotnine import geom_bar

df = pd.DataFrame({
    'x' : ['A','B','C','D'],
    'y' : [1,2,3,4],

})


plot = ggplot(df,aes('x','y')) + geom_bar(stat = 'identity')

plot
```

```python
#Bargraph using Bokeh

output_notebook()


x  = ['A','B','C','D']
y = [1,2,3,4]



plot = figure(title = "Box Graph",x_range = x, x_axis_label = 'x', y_axis_label = 'y')
plot.vbar(x = x,top = y, width = 0.3)

show(plot)
```

```python
#Bargraph using Matplotlib

plt.bar(x,y)
plt.title("Bar Graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Density Plot using Plotnine
from plotnine import geom_density

df = pd.DataFrame({
    'x' : [1,2,2,3,3,3,4,4,4,4],

})


plot = ggplot(df,aes('x')) + geom_density()

plot




```

```python
#Bargraph using Bokeh

from scipy.stats import gaussian_kde

output_notebook()

data = np.random.randn(1000)

density = gaussian_kde(data)

x = np.linspace(-3,3,1000)

y = density(x)

plot = figure(title = "Density Plot",x_axis_label = 'x',y_axis_label = 'y')
plot.line(x, y, line_width = 2)

show(plot)
```

```python
#Density Plot using Matplotlib

data = np.random.randn(1000)

sb.kdeplot(data)
plt.title("Density Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
