```python
#SciPy linar algebra
#SciPy => scientific  python

#Solving Linear equations using scipy linalg

#equations => 7x+2y = 8 , 4x+5y=10
#importing linalg from scipy
from scipy import linalg
import numpy as n

a = n.matrix('7,2;4,5')
b = n.matrix('8;10')

print(a)
print("\n",b)

res = linalg.solve(a,b)
print("\n",res)


```

```python
import matplotlib.pyplot as  plt

x = n.linspace(0,2*n.pi,101)
y = n.sin(x)

#plotting
plt.rc('font',size = 14)
plt.plot(x,y,'black',label = 'sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Scipy quad => to solve integrals
#scipy.integrate.quad module is used
#Note:- Quadrature (quad) requries function f(x)

from scipy.integrate import quad as q

#integral = 3x^2 + 1 from x = 0 to x = 1

def f(x):
  return 3.0*x*x + 1.0

lower_bound = 0
upper_bound = 1

#quad has two variables one is I and the other is error
I_quad , err = q(f,lower_bound,upper_bound)
print(I_quad)
print(err)

#finding only I value
I_quad = q(f,lower_bound,upper_bound)[0]
print(I_quad)

#or

I_quad , _= q(f,lower_bound,upper_bound)
print(I_quad)
```

```python
#Scipy.integrate.simps => solving integrals
#Note:- simps doesnot require a function

#importing simps from scipy.integrate
from scipy.integrate import simps as s

x =  n.arange(0,10)
y = n.arange(0,10)

I_simps = s(x,y)
print(I_simps)
```

```python
#Special functions (scipy.special)
from scipy import special

#Cubic root function
#importing cbrt from scipy.special
from scipy.special import cbrt as c

a = c([27,64])
print(a)

#Exponential function
from scipy.special import exp10 as e

b = e([1,15])
print(b)


#Combinations
from scipy.special import comb as co

d = co(5,2,exact = False , repetition = True)
print(d)


#Permutations
from scipy.special import perm as p

d = p(5,2,exact = True)
print(d)

#log sum exponential function
from scipy.special import logsumexp as lo

x = 1
d = lo(x)
print(d)


#Gamma function
from scipy.special import gamma as g

d = g([0,0.5,1,1,5])
print(d)


#to find sin degree
sin = special.sindg(90)
print(sin)
```

```python
#Symbolic Math
#Sympy Module

#importing sympy
import sympy as sp
sp.init_printing()# to get pretty output

#Note:- use display() instead of print for good output

#Symbol variable

#Syntax

x,y,z = sp.symbols('x,y,z') #Note:- name of the variable doesnot need to be the same as the symbol

ell = sp.symbols('L')

exp1 = x*sp.cos(x)
exp2 = x**2 + y**2 + z
exp3 = sp.sqrt(x+y/2)

#Note:- use display to get the output as symbols
display(x)
display(ell)
display(exp1)
display(exp2)
display(exp3)
```

```python
#Convert strings to sympy expressions

#importing libs
from sympy import *

#String
exp = "x**2 + 3*x -1/2"

#converting into expression
expr = sympify(exp)
display(expr)


#converting expression to string
x,y,z = symbols('x,y,z')
print(str(Integral(sqrt(1/x),x)))
```

```python
#Algebraic operations => simplify , factor , expand

ex = x**2 / x**2
print(sp.simplify(ex))

ex = (x+2)*(x-3)
print(sp.expand(ex))

ex = x**3 - x**2 + x - 1
print(sp.factor(ex))
```

```python
#Evalf => to evaluate a numerical expression into a floating point number

ex = sqrt(8)
ex.evalf()
```

```python
#Substitute a number into an expression
ex = x**3 - x**2 + x - 1
ans1 = ex.subs(x,2.) #evaluates numerically
ans2 = ex.subs(x,2) #treats as a symbol

display(ans1)
display(ans2)

#Substitute a symbol into an expression
ans3 = ex.subs(x,y)
ans4 = ex.subs(x,y**2+z)

display(ans3)
display(ans4)
```

```python
#Calculus

#derivatives are done using sp.diff(expr,var,[order])
#indefinite integrals => sp.integrate(expr,var)
#definite integrals => sp.integrate(expr,(var,lb,ub))

#Examples of derivatives
ex = sp.exp(x**2)
display(sp.diff(ex,x)) #single derivative
display(sp.diff(ex,x,3)) #third  derivative

ex = sp.exp(x*y*z)
dxdy = sp.diff(ex,x,y) #partial derivatives
display(dxdy)
```

```python
#Integrals

display(sp.integrate(x**3,x)) #indefinte integral
display(sp.integrate(x**2,(x,0,1))) #definte integral

display(sp.integrate(sp.exp(-x),(x,0,sp.oo))) #integral with infinity as upper limit
display(sp.integrate(x*sp.exp(-x),(x,y,sp.oo))) #double integral with infinity as limit

```

```python
#Limits

expr = Limit((cos(x)-1)/x,x,0) #to display expression with limit
display(expr)

# doit() => evaluates the expression
display(expr.doit())


#fourier series
ex = x**3
display(sp.series(ex,0,4))

```
