```python
#IPS
def check_password(password):
  if not password[0].isalpha():
    print("Bad")
    return
  if len(password) < 8:
    print("Bad")
    return
  if not any(a.isdigit() for a in password):
    print("Bad")
    return
  if not any(a in "@$!%*?&" for a in password):
    print("Bad")
    return
  print("Good")

password = input()
check_password(password)
```

```python
#IPS
#using regex

import re
def check_password(password):
  if re.match(r"^(?=.*[a-zA-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$", password):
    print("Good")
  else:
    print("Bad")

password = input()

check_password(password)

```
