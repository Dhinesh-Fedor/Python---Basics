```python
#Data Aggregation in Pandas
#Note:= Aggregation is done in column wise


import numpy as np
import pandas as pd


a = np.random.RandomState(42)
b = pd.Series([1,2,3,4])
print(b)

#sum
b.sum()

#min()
print(b.min())

#max
print(b.max())

#mean
print(b.mean())

#first
print(b.first)

#last
print(b.last)

#count
print(b.count())

#standard deviation
print(b.std())

#variance
print(b.var())

#standard error mean
print(b.sem())

#size
print(b.size)

#describe
print(b.describe())
```

```python
#NAN = Not A Number (ie) Null

df = pd.DataFrame([
    [1,2,3],
    [4,5,6],
    [7,8,9],
    [np.nan,np.nan,np.nan]
], columns = ['A','B','C'])


df
```

```python
#Aggregate functions over rows

df.agg(['sum','min','mean','var'])
```

```python
#other way

df.agg({ 'A' : ['sum','max'], 'B' : ['mean','std']})
```

```python
#other way

df.agg(x = ('A','min'), y = ('B','max'))

#x and y are indices (ie) renames the default indices
#Note:- you cant pass a list of aggregate functions as argument (ie) ['min','max] wont work
```

```python
#other way

#axis = 1 => columns, but perfoms row wise aggregation
#axis = 0 => rows, but perfoms column wise aggregation

df.agg("min", axis = 0)
```

```python
#other way

df.mean()
```

```python
#other way

df.mean(axis = 'columns') #1 or 'columns' both works
```

```python
#Groupby : Split, Apply, Combine

df = pd.DataFrame({ 'Key' : ['A','B','C','A','B','C'],
                     'Data' : range(6)
                     })

df
```

```python
#Group by key

df.groupby('Key')
```

```python
#Group by object

dp = pd.DataFrame({
    'Animal' : ['Falcon','Falcon','Parrot'],
    'Speed' : [380,370,24]
})


dp.groupby(['Animal']).mean()
```

```python
#Hierarchical indexes

a = [
    ['Falcon','Falcon','Parrot','Parrot'],
    ['captive','Wild','captive','Wild']
]

i = pd.MultiIndex.from_arrays(a,names = ('Animal','Type'))

dff = pd.DataFrame({'Max Speed' : [390,350,30,20]}, index = i)

dff
```

```python
dff.groupby(level = 0).mean()
```

```python
dff.groupby(level = 'Type').mean()
```

```python
#including None

l = [
    [1,2,3],
    [1,None,4],
    [2,1,3],
    [4,5,6],
]

d = pd.DataFrame(l,columns = ['a','b','c'])

d
```

```python
d.groupby(by = ["b"]).sum()
```

```python
d.groupby(by = ['b'],dropna = False).sum()
```
