```python
#Data Aggregation in Pandas
#Note:= Aggregation is done in column wise


import numpy as np
import pandas as pd


a = np.random.RandomState(42)
b = pd.Series([1,2,3,4])
print(b)

#sum
b.sum()

#min()
print(b.min())

#max
print(b.max())

#mean
print(b.mean())

#first
print(b.first)

#last
print(b.last)

#count
print(b.count())

#standard deviation
print(b.std())

#variance
print(b.var())

#standard error mean
print(b.sem())

#size
print(b.size)

#describe
print(b.describe())
```

```python
#NAN = Not A Number (ie) Null

df = pd.DataFrame([
    [1,2,3],
    [4,5,6],
    [7,8,9],
    [np.nan,np.nan,np.nan]
], columns = ['A','B','C'])


df
```

```python
#Aggregate functions over rows

df.agg(['sum','min','mean','var'])
```

```python
#other way

df.agg({ 'A' : ['sum','max'], 'B' : ['mean','std']})
```

```python
#other way

df.agg(x = ('A','min'), y = ('B','max'))

#x and y are indices (ie) renames the default indices
#Note:- you cant pass a list of aggregate functions as argument (ie) ['min','max] wont work
```
