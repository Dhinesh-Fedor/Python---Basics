```python
#Data Transformation using Pandas

#Pandas DataFrame.transform()

#Syntax => DataFrame.transform(func , axis = 0, *args, **kwargs)

import pandas as pd


df = pd.DataFrame({
    'A' : [12,4,5,None,1],
    'B' : [7,2,54,3,None],
    'C' : [20,16,11,3,8],
    'D' : [14,3,None,2,6],
}, index = range(5))


df
```

```python
res = df.transform(func = lambda x : x + 10)
res
```

```python
res = df.transform(func = ['sqrt','exp'])
res
```

```python
#Other Data Transformation

# 1) Finding Dulplicate values

df1 = pd.DataFrame({
    'a':['one','two']*3,
    'b':[1,1,2,3,2,3]
}, index = range(6))

df1
```

```python
df1.duplicated()
```

```python
#To drop the duplicates

df1.drop_duplicates()
```

```python
df1['c'] = range(6)
df1
```

```python
df1.duplicated(['a','b'], keep = 'last')
```

```python
# 2) Mapping

df2 = pd.DataFrame({
    'names' : ['tim','tom', 'sam', 'kate', 'kim', 'tony'],
    'scores' : [60,50,70,80,40,30]
})

df2
```

```python
classes = {'tim' : 'A', 'tom' : 'A', 'sam' : 'A', 'kate' : 'B', 'kim' : 'B', 'tony' : 'B'}
classes
```

```python
n = df2['names'].str.capitalize()
n
```

```python
df2['branches'] = df2['names'].map(classes)
df2
```

```python
# 3) Replace

s = pd.Series([10,20,30,40])
s
```

```python
import numpy as np

a = s.replace(20, np.nan)
a
```

```python
b = s.replace([10,40,30],[0,0,np.nan])
b
```

```python
c = s.replace({ 20 : np.nan, 30 : 40})
c
```

```python
# 4) Renaming

df4 = pd.DataFrame(np.arange(12).reshape(3,4), columns = ['A','B','C','D'])

s = pd.Series(['one','two','three'])
df4.index = df4.index.map(s)

df4
```

```python
r = df4.rename(index = str.title, columns = str.lower)
r
```

```python
t = df4.rename(index = { 'one' : 1, 'two' : 2, 'three' : 3}, columns = {'A' : 1, 'B' : 2, 'C' : 3, 'D' : 4})
t
```

```python
# 5) cutting

sc = [30,80,40,90,60,45,95,75,55,100,65,85]
sc
```

```python
x = [20,40,60,80,100]
x
```

```python
y =  pd.cut(sc,x)
y
```

```python
y.codes
```

```python
y.categories
```

```python
pd.Series(y).value_counts()
```

```python
z = pd.cut(sc,x,right = False)
z
```

```python
name = ['low','medium','high','very high']
pd.cut(sc,x, labels = name)
```

```python
pd.cut(sc, 10)
```

```python
#qcut => quater cut

data = np.random.randn(100)
c = pd.qcut(data,4)
c
```

```python
pd.Series(c).value_counts()
```
